# How are in-demand skills trending for Data Analysts?
### Methodology
1. Aggregate skill counts monthly
2. Re-analyze based on percentage of total jobs
3. Plot the monthly skill demand


## Importing libraries, loading data and basic data cleanup

```python
#Importing Libraries
import ast
import seaborn as sns
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])  # converting job_posted_date to datetime object
df['job_skills'] = df['job_skills'].apply(lambda data: ast.literal_eval(data) if pd.notna(data) else data)  # converting job_skills to list
```

```python
df_DA_US = df[(df['job_country'] == 'United States') & (df['job_title'] == 'Data Analyst')].copy()  # filter for Data Analysts jobs in US 
```

### 1. Aggregate skill counts monthly

```python
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month  # creating new column for month number 
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')  # exploding job_skills column 
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)  # pivot table

df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()  # add 'Total' row at the end to calculate sum of each skill

df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]  # sort by total skill counts

df_DA_US_pivot = df_DA_US_pivot.drop('Total')  # remove 'Total' row

df_DA_US_pivot
```

### 2. Re-analyze based on percentage of total jobs

```python
DA_month_postings_totals = df_DA_US.groupby('job_posted_month_no').size()  # total count of postings per month
```

```python
df_DA_US_percentage = df_DA_US_pivot.div(DA_month_postings_totals / 100, axis=0)  # Calculate the percentages
```

```python
# Change month number to month name
df_DA_US_percentage = df_DA_US_percentage.reset_index() 
df_DA_US_percentage['job_posted_month'] = df_DA_US_percentage['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percentage = df_DA_US_percentage.set_index('job_posted_month')
df_DA_US_percentage = df_DA_US_percentage.drop(columns='job_posted_month_no')

df_DA_US_percentage
```

### 3. Plot the monthly skill demand

```python
from matplotlib.ticker import PercentFormatter

df_plot = df_DA_US_percentage.iloc[:, :5]  # get top 5 skills

# line plot using seaborn  
sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.despine()

# plot customization
plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()

# format percentages
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

# label each line with skill name
for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i])

plt.show()
```
